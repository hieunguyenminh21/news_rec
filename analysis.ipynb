```python
# plot = df[(df['history_user_encoder_method'] == 'MultiHeadSelfAttention')] \
plot = df[(df['raw_text_encoder_method'] == 'CNN') & (df['history_user_encoder_method'] == 'Transformer')] \
[['raw_text_encoder_method', 'history_user_encoder_method', 'text_encoder_method', 'user_history_augmentation_method', \
  'recommend_weight', 'AUC', 'MRR', 'NDCG@5', 'NDCG@10']]
```

```python
plot = plot.sort_values('recommend_weight')
```

```python
plot = plot[(plot['user_history_augmentation_method'] == 'Version2')]
```

```python
plot.sort_values('AUC', ascending=False)
```

```python
rec_weight = plot['recommend_weight'].tolist()
```

```python
auc = plot['AUC'].tolist()
mrr = plot['MRR'].tolist()
ndcg5 = plot['NDCG@5'].tolist()
ndcg10 = plot['NDCG@10'].tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy as np

xticks = rec_weight
fig, ax = plt.subplots()

ax.plot(rec_weight, auc, label = 'auc')
ax.plot(rec_weight, mrr, label = 'mrr')
ax.plot(rec_weight, ndcg5, label = 'ndcg@5')
ax.plot(rec_weight, ndcg10, label = 'ndcg@10')
ax.set(xticks=xticks, xticklabels=xticks)


ax.legend(loc=0)
plt.xlabel('Lambda')
plt.show()
```
